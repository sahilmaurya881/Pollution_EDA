```python
import pandas as pd
import numpy as np
```

```python
deff=pd.read_csv('/content/pollution_data.csv',)
df=deff
deff
```

```python
from datetime import datetime
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
del_cols = ['State','mil_miles', 'past_week_avg_miles' , 'latitude','longitude']
df = df.drop(del_cols, axis=1)
df
```

```python
o3_name = ['Date','City','County','o3_count','o3_min','o3_median','o3_max','o3_variance']
DataO3 = df[o3_name]
DataO3.describe()
```

```python
DataO3 = DataO3.fillna(DataO3.mean().round(2))
DataO3
```

```python
pressure_name=['Date','City','County','pressure_count','pressure_min','pressure_median','pressure_max','pressure_variance']
Data_pressure=df[pressure_name]
Data_pressure.describe()
```

```python
Data_pressure = Data_pressure.fillna(Data_pressure.mean().round(2))
Data_pressure.describe()
```

```python
pm_name=['Date','City','County','pm25_count','pm25_min','pm25_median','pm25_max','pm25_variance']
Data_pm=df[pm_name]
Data_pm.describe()
```

```python
Data_pm = Data_pm.fillna(Data_pm.mean().round(2))
Data_pm.describe()
```

```python
humidity_name=['Date','City','County','humidity_count','humidity_min','humidity_median','humidity_max','humidity_variance']
Data_humidity=df[humidity_name]
Data_humidity.describe()
```

```python
Data_humidity = Data_humidity.fillna(Data_humidity.mean().round(2))
Data_humidity.describe()
```

```python
temperature_name=['Date','City','County','temperature_count','temperature_min','temperature_median','temperature_max','temperature_variance']
Data_temperature=df[temperature_name]
Data_temperature.describe()
```

```python
Data_temperature = Data_temperature.fillna(Data_temperature.mean().round(2))
Data_temperature.describe()
```

```python
wind_speed_name=['Date','City','County','wind-speed_count','wind-speed_min','wind-speed_median','wind-speed_max','wind-speed_variance']
Data_wind_speed=df[wind_speed_name]
Data_wind_speed.describe()
```

```python
Data_wind_speed = Data_wind_speed.fillna(Data_wind_speed.mean().round(2))
Data_wind_speed.describe()
```

```python
wind_gust_name=['Date','City','County','wind-gust_count','wind-gust_min','wind-gust_median','wind-gust_max','wind-gust_variance']
Data_wind_gust=df[wind_gust_name]
Data_wind_gust.describe()
```

```python
Data_wind_gust = Data_wind_gust.fillna(Data_wind_gust.mean().round(2))
Data_wind_gust.describe()
```

```python
del_cols = ['Date','City','County']
Data_pm = Data_pm.drop(del_cols, axis=1)
Data_wind_gust = Data_wind_gust.drop(del_cols, axis=1)
Data_wind_speed = Data_wind_speed.drop(del_cols, axis=1)
Data_temperature = Data_temperature.drop(del_cols, axis=1)
Data_humidity = Data_humidity.drop(del_cols, axis=1)
Data_pressure = Data_pressure.drop(del_cols, axis=1)
```

```python
result = DataO3.join(Data_pm, lsuffix='DataO3', rsuffix='Data_pm', how='inner')
result = result.join(Data_wind_gust, lsuffix='', rsuffix='Data_wind_gust', how='inner')
result = result.join(Data_wind_speed, lsuffix='', rsuffix='Data_wind_speed', how='inner')
result = result.join(Data_temperature, lsuffix='', rsuffix='Data_temperature', how='inner')
result = result.join(Data_humidity, lsuffix='', rsuffix='Data_humidity', how='inner')
result = result.join(Data_pressure, lsuffix='', rsuffix='Data_pressure', how='inner')
```
